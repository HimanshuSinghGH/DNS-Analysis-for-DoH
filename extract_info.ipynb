```python
from scapy.utils import RawPcapReader
import pandas as pd
import numpy as np
import os
```

## Function to open pcap file and extract count of total packets

```python
def process_pcap(file_name):
    print('Opening {}...'.format(file_name))

    count = 0
    for (pkt_data, pkt_metadata,) in RawPcapReader(file_name):
        count += 1

    return count
```

```python
df = pd.read_csv('top-1m.csv' , names=['rank','website'])
```

```python
site_df = df.iloc[500:650]

site_df.insert(2, "raw_dns", int(0)) 

print(site_df)

```

## Open folder containing pcap files and append info in DataFrame

```python
#for (root,dirs,files) in os.walk('./capturefiles', topdown=False): 
for root, dirs, files in os.walk('./uncaptured'):
    for f in files :
        print(f)
        file_path = './uncaptured/'+f
        #print(file_path)
        dns = process_pcap(file_path)
        rank = f[4:]    
        rank = rank[:-5]
        rank = int(rank) 
        dns = int(dns)
        site_df.at[rank-1 , 'raw_dns'] = dns


        
```
